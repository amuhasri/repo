```python
df = pd.merge(df1, df2, left_on='menu_item_id', right_on='item_id', how='left')
df
```

```python
#Drop 'item_id' column, because its = 'menu_item_id'
df = df.drop(columns=['item_id'])
```

```python
# Create function am-pm to 24 hour
def convert_to_24_hour(time_str):
    # Convert to datetime object with AM/PM format
    dt_obj = datetime.strptime(time_str, '%I:%M:%S %p')
    # Format the datetime object to 24-hour format
    return dt_obj.strftime('%H:%M:%S')
```

```python
# Extract the time part in 24-hour format
df['order_time24'] = df['order_time'].apply(convert_to_24_hour)
df.head()
```

```python
# Drop 'order_time'
df = df.drop(columns=['order_time'])
df.head()
```

```python
# Makesure Null Values
df.isna().sum()
```

```python
df.describe()
```

```python
#Find avg value to make sure Standard Deviation is valid (std < avg)
df.mean()
```

```python
#Change datatype for 'order_date' column
df['order_date'] = pd.to_datetime(df['order_date'])
```
