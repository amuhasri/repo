### Read Data

```python
#Show Data
df = pd.read_csv('/content/ORI Retail_Transactions_Dataset.csv')
df.head()
# df.head(new_df = original_df['OriginalColumn'].str.split(', ', expand=True))
```

### Cleaning Data

```python
#Cleaning Data, remove [] and '' on product value column
df = df.replace('[\[\]]', '', regex=True)
df = df.replace('[\'\']', '', regex=True)
df.head()
```

```python
#Split Product Column in new Table named Product_df
Product_df = df['Product'].str.split(', ', expand=True)
Product_df.head()
```

```python
#Add Transaction_ID to Product_df
Product_df['Transaction_ID'] = df['Transaction_ID']
Product_df.head()
```

```python
# Make Product_df into 2 column only, Transaction_ID & Product_Purchase
melted_df = pd.melt(Product_df, id_vars=['Transaction_ID'], value_vars=[0, 1, 2, 3, 4], value_name='Product_Purchase')
melted_df.head
```

```python
#Melt all product into one column
melted_df = melted_df.drop(columns=['variable'])
melted_df = melted_df.dropna(subset=['Product_Purchase'])
melted_df = melted_df.reset_index(drop=True)
melted_df
```

```python
#Checking Transaction_ID 1000000001 (make sure is the list of product correct)
melted_df[melted_df.Transaction_ID	==1000000001]
```

```python
# Spliting timestamp
def get_hour(x):
  return int(x.split(' ')[1].split(':')[0])

df['hour']=df.Date.apply(get_hour)
df.head()
```

```python
#Change datatype for 'date' column
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#seperate date into detail
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Minute'] = df['Date'].dt.minute
df.head()
```

```python
#Drop product column, because its not necessary in table
df.drop(columns=['Product'], inplace=True)
df.head()
```

```python
# Joining df with melted.df
df_new = pd.merge(df, melted_df, on='Transaction_ID', how='inner')
df_new
```

```python
#Duplicate Product_Purchase before one-hot encode
df_new['ProductPurchase'] = df_new['Product_Purchase']
df_new.head()
```

```python
#Change datatype for Total Cost into float
df_new['Total_Cost'] = df_new['Total_Cost'].astype(float)
```

```python
#Add new column of One Hot Encoding for Product_Purchase
df_new = pd.get_dummies(df_new, columns=['ProductPurchase'], dtype=int)
df_new
```

```python
#Make sure Product_Purchase column still on table
df_new['Product_Purchase'].unique()
```

```python
#Checking if there any NA value on table
df_new.isna().sum()
```
